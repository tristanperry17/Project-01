```python
import pandas as pd
import matplotlib.pyplot as plt

# Study data files
beaufortpath = "Beaufort.csv"

beaufortdata = pd.read_csv(beaufortpath)

# Specify the columns for the x and y axes
x_column = 'DATE'
y_column = 'PRCP'

# Plot the line graph
plt.plot(beaufortdata[x_column], beaufortdata[y_column])

# Add labels and title
plt.xlabel('DATE')
plt.ylabel('PRCP')
plt.title('PRCP over time'.format(y_column, x_column))

# Display the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.dates import YearLocator

# Study data files
beaufortpath = "Beaufort.csv"

bf = pd.read_csv(beaufortpath)

# Specify the columns for the x and y axes
x_column = 'DATE'
y_column = 'PRCP'

bf[x_column] = pd.to_datetime(bf[x_column])

# Set the size of the figure
plt.figure(figsize=(12, 6))  # Adjust the width (12) and height (6) as needed

# Plot the line graph with a dashed line style
plt.plot(bf[x_column], bf[y_column], linestyle='--', label='Precipitation in CM')

# Add labels and title
plt.xlabel('Date')
plt.ylabel('PRCP')
plt.title('PRCP over time'.format(y_column, x_column))

# Set x-axis ticks to display every year
plt.gca().xaxis.set_major_locator(YearLocator())

plt.xticks(rotation=45)


plt.legend()

# Display the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.dates import YearLocator

# Study data files
manteopath = "Manteo.csv"
mt = pd.read_csv(manteopath)

# Specify the columns for the x and y axes
x_column = 'Date'
y_column = 'PRCP'

mt[x_column] = pd.to_datetime(mt[x_column])

# Set the size of the figure
plt.figure(figsize=(12, 6))

# Plot the line graph with a dashed line style
plt.plot(mt[x_column], mt[y_column], linestyle='--', label='Precipitation in Inches')

# Add labels and title
plt.xlabel('DATE')
plt.ylabel('PRCP')
plt.title('PRCP over time'.format(y_column, x_column))

# Set x-axis ticks to display every year
plt.gca().xaxis.set_major_locator(YearLocator())

plt.xticks(rotation=45)

plt.legend()

# Display the plot
plt.show()
```
